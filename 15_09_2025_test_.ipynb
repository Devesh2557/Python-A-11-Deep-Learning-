```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/Attrition.csv')
```

```python
df1 = df.drop(columns = ['BusinessTravel'])
```

```python
df1.dtypes
```

```python
# df.isnull().sum()
```

```python
x = df1.drop(columns = 'Attrition')
y = df['Attrition']
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=42)
```

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer = ColumnTransformer(transformers=[
    ('tf1',StandardScaler(),['Age','DailyRate', 'DistanceFromHome', 'Education', 'EmployeeCount', 'EmployeeNumber', 'EnvironmentSatisfaction','HourlyRate',
                             'JobInvolvement','JobLevel', 'JobSatisfaction', 'MonthlyIncome', 'MonthlyRate', 'NumCompaniesWorked','PercentSalaryHike',
                             'PerformanceRating', 'RelationshipSatisfaction', 'StandardHours', 'StockOptionLevel', 'TotalWorkingYears', 'TrainingTimesLastYear',
                             'WorkLifeBalance','YearsAtCompany','YearsInCurrentRole','YearsSinceLastPromotion','YearsWithCurrManager']),

    ('tf2', OneHotEncoder(sparse_output = True , drop = 'first'), ['Department', 'EducationField', 'Gender','JobRole','MaritalStatus', 'Over18',
                                                                    'OverTime'])
], remainder = 'passthrough')
```

```python
transformer.fit_transform(x_train).shape
```

```python
transformer.fit_transform(x_test).shape
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
```

```python
rf_pi = Pipeline(steps=[
    ('transformer', transformer),
    ('classifier', RandomForestClassifier(random_state=42))
])
```

```python
rf_pi.fit(x_train, y_train)
```

```python
y_pred = rf_pi.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test , y_pred)
```
